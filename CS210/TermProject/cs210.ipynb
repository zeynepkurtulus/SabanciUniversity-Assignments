#GEOGRAPHİCAL COMPARISON OF CONFIRMED, DEATHS AND ACTIVE CASES

```python
 df_aggre = df.groupby(['Lat',"Long"]).agg({'Confirmed' : 'sum', 'Deaths': 'sum', 'Recovered': 'sum' , 'Active': 'sum'}).reset_index()
```

```python
r = 1
c = 2
fig, axes = plt.subplots(r, c, sharex=True, figsize=(15, 5))
ax = axes[0]
df_aggre.plot(x = 'Lat', y='Deaths', rot=90, ax = ax)
ax = axes[1]
ax = df_aggre.plot(x = 'Lat', y='Confirmed', rot=90, ax =ax)


```

```python
df_aggre = df.groupby(["Long", 'Lat']).agg({'Confirmed' : 'sum', 'Deaths': 'sum', 'Recovered': 'sum' , 'Active': 'sum'}).reset_index()
```

```python
r = 1
c = 2
fig, axes = plt.subplots(r, c, sharex=True, figsize=(15, 5))
ax = axes[0]
df_aggre.plot(x = 'Long', y='Deaths', rot=90, ax = ax)
ax = axes[1]
ax = df_aggre.plot(x = 'Long', y='Confirmed', rot=90, ax=ax)
```

We can see from the graphs that if one countries number of deaths or confirmed cases are high, the nearby countries tend to be high too because of the spread of the virus. We clearly see from both sections that if one countrie's either latitude or longitude or both are close to another countries, they more likely have similar confirmed and death cases.
